```python
!mamba install pandas
import pandas as pd

df = pd.read_csv("reddit_posts.csv")

df.head()
```

```python
df["engagement"] = df["score"] + df["num_comments"]

df[["score","num_comments","engagement"]].head()
```

```python
emotion_stats = (
    df.groupby("dominant_emotion")["engagement"]
      .mean()
      .sort_values(ascending=False)
)

print(emotion_stats)
```

```python
import matplotlib.pyplot as plt

emotion_stats.plot(kind="bar")

plt.title("Average Engagement by Emotion")
plt.ylabel("Average Engagement")
plt.show()
```

```python
subreddit_stats = (
    df.groupby("subreddit")["engagement"]
      .mean()
      .sort_values(ascending=False)
)

print(subreddit_stats)
```

```python
subreddit_stats.plot(kind="bar")

plt.title("Average Engagement by Subreddit")
plt.ylabel("Engagement")
plt.show()
```

```python
df["dominant_emotion"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Emotion Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("reddit_posts.csv")

df["engagement"] = df["score"] + df["num_comments"]

# Emotion Distribution
plt.figure(figsize=(8,5))
df["dominant_emotion"].value_counts().plot(kind="bar")
plt.title("Emotion Distribution")
plt.tight_layout()
plt.savefig("emotion_distribution.png")
plt.close()

# Engagement by Emotion
plt.figure(figsize=(8,5))
df.groupby("dominant_emotion")["engagement"].mean().sort_values().plot(kind="bar")
plt.title("Average Engagement by Emotion")
plt.tight_layout()
plt.savefig("engagement_by_emotion.png")
plt.close()

# Subreddit Performance
plt.figure(figsize=(8,5))
df.groupby("subreddit")["engagement"].mean().sort_values().plot(kind="bar")
plt.title("Average Engagement by Subreddit")
plt.tight_layout()
plt.savefig("subreddit_performance.png")
plt.close()

print("Charts created")
```

```python
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    PageBreak,
    Image
)

from reportlab.lib.styles import getSampleStyleSheet

import pandas as pd

df = pd.read_csv("reddit_posts.csv")

df["engagement"] = df["score"] + df["num_comments"]

styles = getSampleStyleSheet()

report = SimpleDocTemplate(
    "GenAI_Social_Intelligence_Report.pdf"
)

elements = []

# -----------------
# PAGE 1
# -----------------

elements.append(
    Paragraph(
        "GenAI Social Media Intelligence Report",
        styles['Title']
    )
)

elements.append(Spacer(1,12))

total_posts = len(df)
avg_engagement = round(df["engagement"].mean(),2)

top_emotion = (
    df.groupby("dominant_emotion")["engagement"]
    .mean()
    .idxmax()
)

elements.append(
    Paragraph(
        f"""
        <b>Executive Summary</b><br/><br/>
        Total Posts Analysed: {total_posts}<br/>
        Average Engagement: {avg_engagement}<br/>
        Highest Performing Emotion: {top_emotion}<br/><br/>

        This report analyzes Reddit discussions across
        multiple communities to uncover emotional patterns,
        engagement drivers and content trends.
        """,
        styles['BodyText']
    )
)

elements.append(PageBreak())

# -----------------
# PAGE 2
# -----------------

elements.append(
    Paragraph(
        "Emotion Intelligence Analysis",
        styles['Heading1']
    )
)

elements.append(
    Image(
        "emotion_distribution.png",
        width=450,
        height=250
    )
)

elements.append(
    Image(
        "engagement_by_emotion.png",
        width=450,
        height=250
    )
)

elements.append(PageBreak())

# -----------------
# PAGE 3
# -----------------

elements.append(
    Paragraph(
        "Community Performance Analysis",
        styles['Heading1']
    )
)

elements.append(
    Image(
        "subreddit_performance.png",
        width=450,
        height=250
    )
)

top_subreddit = (
    df.groupby("subreddit")["engagement"]
    .mean()
    .idxmax()
)

elements.append(
    Paragraph(
        f"""
        Highest Performing Community:
        <b>{top_subreddit}</b>
        """,
        styles['BodyText']
    )
)

elements.append(PageBreak())

# -----------------
# PAGE 4
# -----------------

elements.append(
    Paragraph(
        "Top Content Analysis",
        styles['Heading1']
    )
)

top_posts = (
    df.sort_values(
        "engagement",
        ascending=False
    )
    .head(10)
)

for i,row in top_posts.iterrows():

    elements.append(
        Paragraph(
            f"""
            <b>{row['title']}</b><br/>
            Engagement: {row['engagement']}
            """,
            styles['BodyText']
        )
    )

elements.append(PageBreak())

# -----------------
# PAGE 5
# -----------------

elements.append(
    Paragraph(
        "Strategic Recommendations",
        styles['Heading1']
    )
)

elements.append(
    Paragraph(
        f"""
        1. Content associated with {top_emotion}
        emotions generates the strongest engagement.

        <br/><br/>

        2. Communities similar to {top_subreddit}
        should receive greater content focus.

        <br/><br/>

        3. Emotional resonance appears to be a stronger
        predictor of engagement than posting volume.

        <br/><br/>

        4. Future work can include real-time Reddit
        monitoring and LLM-generated insight reports.
        """,
        styles['BodyText']
    )
)

report.build(elements)

print("Professional report created successfully.")
```
